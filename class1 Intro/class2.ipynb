## Types of variables

#different types of variables are

        #integer(int)--> whole num
        #eg: age = 20

        #float--> Decimal num
        #eg: price = 9.5

        #string (str) --> text/word
        #eg: name = "areesha raza"

        #boolean(bool)--> true/false
        #eg: is_student = True

        #list--> Multiple values 
        #eg: num =[1,2,3,4...]

        #Tuples --> like list but immutable
        #eg: colors= ("red", "yellow"....)

        #Dictionary--> key_value pair 
        #eg: students= {"name":"Ali","age":23..}
